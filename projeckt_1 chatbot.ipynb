Das Ziel ist, Website-Besucher/innen zu grüßen, ihnen eine BMI-Rechnung anbieten, ihre Informationen wie Alter, Größe und Gewicht einzusammeln und ihr BMI zu berechnen und ausgeben.

Was ist der BMI?
Der Body-Mass-Index, kurz BMI, ist die gebräuchlichste Formel zur Gewichtsberechnung. Er ergibt sich aus dem Verhältnis des Körpergewichts in Kilogramm und der Körpergröße in Metern zum Quadrat. Je nach Höhe des errechneten Werts unterscheidet die Deutsche Gesellschaft für Ernährung (DGE) für die Auswertung fünf Kategorien: Untergewicht, Normalgewicht, Übergewicht, extremes Übergewicht (Adipositas) und massive Adipositas.

BMI = Körpergewicht : (Körpergröße)²

Ein Beispiel: Ein 30-jähriger Mann mit einem Körpergewicht von 80 Kilogramm hat bei einer Körpergröße von 1,85 Metern einen BMI von 23,37 und ist damit normalgewichtig.

Gewicht-Kategorien:

Kritisches Untergewicht wenn bmi < 17.5
Untergewicht wenn bmi zwischen 17.5 und 22
Normalgewicht wenn bmi zwischen 22 und 28
Leichtes Übergewicht wenn bmi zwischen 28 und 33
Übergewicht wenn bmi größer als 33
Für jede Kategorie soll der Chatbot den BMI-Wert zusammen mit folgenden Nachrichten ausgeben:

Achtung: Ein BMI-Wert unter 17,5 ist bedenklich und gilt bei Männern wie Frauen als anorektisch. Um mit einem viel zu niedrigen Körpergewicht nicht die Gesundheit zu gefährden, sollten Sie bei einem solchen Ergebnis dringend einen Arzt aufsuchen.
Ihr BMI ist sehr niedrig. Versuchen Sie, etwas zuzunehmen, um langfristig gesund und leistungsfähig zu bleiben. Die Techniker oder Ihr Arzt unterstützen Sie gern dabei.
Super! Ihr Gewicht ist gesund. Mit einer ausgewogenen Ernährung und regelmäßiger Bewegung wird das auch langfristig so bleiben.
Ihr BMI ist leicht erhöht. Achten Sie auf Ihre Ernährung und bewegen Sie sich regelmäßig. So können Sie das Risiko für mögliche Begleiterkrankungen minimieren.
Ihr BMI weist auf Übergewicht hin. Das birgt die Gefahr von Krankheiten wie etwa Diabetes. Ihr Hausarzt und die Techniker unterstützen Sie gern bei einem gesunden Lebensstil.

```python
intentions = { 'intents':[
    { 
      'tag': 'greetings',
      'pattern':['hallo', 'hi', 'ja','hey', 'Guten Tag', 'guten morgen', 'guten abend'],
      'responses':['Hallo.', 'Wie kann ich helfen?', 'Haben Sie Fragen?', 'Sie können mich gerne etwas fragen.']
    },

    {
        'tag':'farewell',
        'pattern':['tschüss','nichts','nein danke','nein', 'ciao', 'auf wiedersehen', 'auf wiederhören', 'bis dann', 'bye','bye bye', 'schönen tag noch', 'schönen abend noch'],
        'responses':['Danke für Ihre Besuch. Auf Wiedersehen']
    },
    {
        'tag':'grateful',
        'pattern':['super danke','danke','super','super,danke','gut,danke','gut','ok','alles gut'],
        'responses':['Gerne, was kann ich noch helfen?','Gerne, haben Sie noch Fragen?']
    },
    {
        'tag':'quastion',
        'pattern':['was kannst du ?','wie heißt du?','wie alte bist du?','was ist dein Geschlecht?','hast du freunde?','wie schlau bist du?'],
        'responses':['das sage ich nicht,haben Sie noch Fragen?','das ist ein Geheimnis, haben Sie noch Fragen?']  
    },
    {
        'tag':'noanswer',
        'pattern':[''],
        'responses':['Sorry, ich habe nicht verstanden','Bitte geben Sie mir mehr Info','Bitte sagen Sie ein bisschen mehr']    
    },
    {
        'tage':'appointment',
        'pattern':['kann ich ein Termin online machen?','ich möchte ein Termin machen','wie man ein Termin machen kann?'],
        'responses':['Sie können gerne dieses Nummer anfrufen:030-1234567']
    },
    {
       'tag':'time',
       'pattern':['wann haben Sie geöffnet?', 'öffnungszeiten', 'wann haben sie geöffnet'],
       'responses':['unsere online akademie ist 24/7 geöffnet!', '24/7']
    }]
}
```

```python
inten_list = intentions.setdefault('intents')
inten_list
```

```python
import random
import re
def bmi(): 
   
    a = 'Achtung: Ein BMI-Wert unter 17,5 ist bedenklich und gilt bei Männern wie Frauen als anorektisch. Um mit einem viel zu niedrigen Körpergewicht nicht die Gesundheit zu gefährden, sollten Sie bei einem solchen Ergebnis dringend einen Arzt aufsuchen.'
    b = 'Ihr BMI ist sehr niedrig. Versuchen Sie, etwas zuzunehmen, um langfristig gesund und leistungsfähig zu bleiben. Die Techniker oder Ihr Arzt unterstützen Sie gern dabei.'
    c = 'Super! Ihr Gewicht ist gesund. Mit einer ausgewogenen Ernährung und regelmäßiger Bewegung wird das auch langfristig so bleiben.'
    d = 'Ihr BMI ist leicht erhöht. Achten Sie auf Ihre Ernährung und bewegen Sie sich regelmäßig. So können Sie das Risiko für mögliche Begleiterkrankungen minimieren.'
    e = 'Ihr BMI weist auf Übergewicht hin. Das birgt die Gefahr von Krankheiten wie etwa Diabetes. Ihr Hausarzt und die Techniker unterstützen Sie gern bei einem gesunden Lebensstil.'
    print('Bot-->ich antworte gerne, ich brauche nur ein paar Fragen.')
    while True:
        weight = ''
        try:
            w=float(input('Bot-->Bitte geben Sie Ihre Gewicht ein (in Kilos): \n'))
            if 500 > w > 5:
                weight=w
                break
            else: 
                print('Bot-->BITTE GEBEN SIE EINE GÜLTIGE NUMMER EIN!' )
                continue
        except ValueError: 
            print ('Bot-->BITTE GEBEN SIE DAS RICHTIGE FORMAT EIN!' )
            continue
    print('Bot-->sehr gut!Jetzt nächste Frage.')    
    while True:  
        height = ''
        try:
            h=float(input('Bot--> Bitte geben Sie hier Ihre Körpergröße ein (in Meter): \n'))           
            if 3>h>0.5:
                height=h
                break
                
            else:
                print('Bot-->BITTE GEBEN SIE EINE GÜLTIGE NUMMER EIN!!')
                continue
        except ValueError: 
            print ('Bot-->BITTE GEBEN SIE DAS RICHTIGE FORMAT EIN!')
            continue
        break

    bmi =  round(float(weight / (height**2)),2)
    print('Bot--> Danke! Ihr BMI ist: '+str(bmi))
    if bmi< 17.5:
        return a
    elif  17.5 < bmi < 22:
        return  b
    elif  22 < bmi < 28:
        return c
    elif 28 < bmi < 33:
        return d
    else:
        return e
    

def bot_response():
    uz = input("You-->").lower()
    
    for i in range(len(inten_list)):
        antw = False
        x = inten_list[i].get('pattern')
        if uz in x:
            antw = True
            return str("Bot-->" + random.choice(inten_list[i].get('responses')))  
    
    if antw == False:
        mylist=['bmi']
        for i in uz:
            i=re.split(r'\s',uz.lower())
            for word in i:
                if word in mylist:
                    return bmi()
        return "Sorry, ich bin noch dabei zu lernen, Sie können mich gerne zum BMI fragen"
    
def loop():
    print('Hallo, willkommen bei sixpaxx Fitnessstudio, ich bin Ihr BMI Berater')
    while True:
        u = bot_response()
        print(u)
        if u == "Bot-->" + random.choice(inten_list[1].get('responses')): 
            break
        
loop()
```
